```python
sn.scatterplot(data=df, x='households', y='population')
```

```python
from seaborn.utils import load_dataset
penguins = sn.load_dataset('penguins')
penguins.head()
```

Создаем dataset для работы с penguins

```python
sn.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='sex', style='island', size=5)
```

```python
list_p = ['species', 'island',	'bill_length_mm',	'bill_depth_mm',	'flipper_length_mm',	'body_mass_g',	'sex']
pg = sn.PairGrid(penguins[list_p])
pg.map(sn.scatterplot)
```

```python
sn.histplot(data=penguins, x='species')
```

```python
sn.histplot(data=penguins, x='island')
```

```python
sn.histplot(data=penguins, x='bill_length_mm')
```

```python
test = penguins.head(3).pivot('bill_length_mm',	'bill_depth_mm',	'flipper_length_mm')
sn = sn.heatmap(test, annot=True)
```

```python
penguins.loc[penguins['bill_length_mm'] >= 42, 'high_group'] = 'high'
penguins.loc[(penguins['bill_length_mm'] >= 35) & (penguins['bill_length_mm'] < 42), 'high_group'] = 'middle'
penguins.loc[(penguins['bill_length_mm'] < 35), 'high_group'] = 'low'
penguins.head()
```

Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина (high - длинный, от 42 мм; middle - средний, от 35 до 42 мм; low - маленький, до 35 мм)

```python
sn.displot(penguins, x='flipper_length_mm', hue='high_group')
```

```python
sn.displot(penguins, x='flipper_length_mm', hue='high_group', multiple='stack')
```
